```python
from ultralytics import YOLO
import cv2

# Load model  hun luyn
model = YOLO("best(1).pt")  # ng dn n best.pt sau khi train

# M webcam (0 = webcam mc nh)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # D on vi YOLO
    results = model.predict(source=frame, imgsz=640, conf=0.4, show=False)

    # Hin th kt qu
    annotated_frame = results[0].plot()
    cv2.imshow("YOLOv8 - Webcam Detection", annotated_frame)

    # Nhn 'q'  thot
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Ti m hnh YOLOv8
model = YOLO("best_model.pt")

# S dng lung video t camera in thoi qua DroidCam
cap = cv2.VideoCapture("http://192.168.1.83:4747/video")  # a ch t nh bn gi

if not cap.isOpened():
    print("Khng th kt ni n camera in thoi")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Khng nhn c frame")
        break

    # Nhn dng bng YOLO
    results = model(frame)

    # V kt qu nhn dng
    annotated_frame = results[0].plot()

    # Hin th kt qu
    cv2.imshow("YOLOv8 - Phone Camera", annotated_frame)

    # Nhn 'q'  thot
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
