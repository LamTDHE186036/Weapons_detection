```python
from ultralytics import YOLO
import cv2

# Load model đã huấn luyện
model = YOLO("best(1).pt")  # Đường dẫn đến best.pt sau khi train

# Mở webcam (0 = webcam mặc định)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Dự đoán với YOLO
    results = model.predict(source=frame, imgsz=640, conf=0.4, show=False)

    # Hiển thị kết quả
    annotated_frame = results[0].plot()
    cv2.imshow("YOLOv8 - Webcam Detection", annotated_frame)

    # Nhấn 'q' để thoát
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Tải mô hình YOLOv8
model = YOLO("best_model.pt")

# Sử dụng luồng video từ camera điện thoại qua DroidCam
cap = cv2.VideoCapture("http://192.168.1.83:4747/video")  # địa chỉ từ ảnh bạn gửi

if not cap.isOpened():
    print("Không thể kết nối đến camera điện thoại")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Không nhận được frame")
        break

    # Nhận dạng bằng YOLO
    results = model(frame)

    # Vẽ kết quả nhận dạng
    annotated_frame = results[0].plot()

    # Hiển thị kết quả
    cv2.imshow("YOLOv8 - Phone Camera", annotated_frame)

    # Nhấn 'q' để thoát
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
